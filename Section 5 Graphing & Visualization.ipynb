# 48. Plotting coordinates on a plane

```python
import matplotlib.pyplot as plt
```

```python
x = 2
y = 0

#ni kedudukan titik
plt.plot(x,y,'gs')

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()

plt.show()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]


#ni kedudukan titik
for i in range(0, len(x)):
    plt.plot(x[i],y[i],'o', label='point %s' %i)

#ni shape graph
plt.axis('square')

plt.grid()
plt.legend()

plt.show()
```

```python
plt.plot(4,2,'r+')

axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0],5.4854854568])

plt.show()
```

# 49. Plotting coordinates on a plane: exersice

```python
import matplotlib.pyplot as plt

x = list(range(-10,12))

#kedudukan point
for i in x:
    j = i**2 - 3*i
    plt.plot(i,j, 'o')
    
#bagi grid
plt.grid()

plt.xlabel('x')
plt.ylabel('f(x) = x**2 - 3*x')
    
    
    
```

### Sympy in matplotlib
Kali ni nak tunjukan kenapa, Mike beria nak ajar guna Sympy ni, sbb dengan cara orang senang nak faham, first kene declare yang X tu jadikan sebagai symbols. Lepas tu baru bleh pakai dengan sympy ni.

```python
import sympy as sym

# kene declare nak pakai X ni as symbols
x = sym.symbols('x')
y = x**2 - 3*x

for i in range(-11,12):
    plt.plot(i,y.subs(x,i), 'o')
```
