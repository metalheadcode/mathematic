```python
import matplotlib.pyplot as plt

x = list(range(-10,12))

#kedudukan point
for i in x:
    j = i**2 - 3*i
    plt.plot(i,j, 'o')
    
#bagi grid
plt.grid()

plt.xlabel('x')
plt.ylabel('f(x) = x**2 - 3*x')
    
    
    
```

### Sympy in matplotlib
Kali ni nak tunjukan kenapa, Mike beria nak ajar guna Sympy ni, sbb dengan cara orang senang nak faham, first kene declare yang X tu jadikan sebagai symbols. Lepas tu baru bleh pakai dengan sympy ni.

```python
import sympy as sym
import matplotlib.pyplot as plt

# kene declare nak pakai X ni as symbols
x = sym.symbols('x')
y = x**2 - 3*x

for i in range(-11,12):
    plt.plot(i,y.subs(x,i), 'o')
    
plt.xlabel('x')
plt.ylabel('y = %s' %y)

plt.show()
```

# 51. Graphing Line Part 2

```python
#y=mx+b
import matplotlib.pyplot as plt

x = [-5,5]
m = -.3
b = .1

y = [340,2345]

for i in range(0, len(x)):
    y[i] = m*x[i] +b
    
plt.plot(x,y,'b--',label='y=%sx+%s' %(m,b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0], 'k')
plt.plot([0,0],axis.get_xlim(), 'k')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

m = [0.7,-1.25]
b = [-2,0.75]
x = [-5,5]

for i in range(2):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y, label = 'y = %sx + %s' %(m[i],b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

# declare axis
# gca stand for get_current_axis
axis = plt.gca()
# melintang
plt.plot(axis.get_xlim(),[0,0], 'k--')
# menegak
plt.plot([0,0], axis.get_xlim(), 'k--')

# xlable
plt.xlabel('x')
# ylabel
plt.ylabel('y')
# title
plt.title('The plot')

plt.legend()
    
plt.show()
```

# 52. Graphing rational functions

Rational ni bermaksud...

```python
import matplotlib.pyplot as plt
import numpy as np

x = range(-3,4)
y = np.zeros(len(x))


for i in range(0, len(x)):
    y[i] = 2 - x[i] ** 2
    
print(x,y)

plt.plot(x,y,'s-')
plt.show()
    
```
